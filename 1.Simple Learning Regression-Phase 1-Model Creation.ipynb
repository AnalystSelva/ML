```python
# Library to load the excel,db,csv file and stored as datframe
import pandas as pd
```

```python
# load the csv file as dataframe  in dataset variable
dataset = pd.read_csv("Salary_Data.csv")
```

```python
# check the result in dataset variable
dataset
```

```python
# split input and output data
# select the independent column from the dataframe
independent = dataset[['YearsExperience']]
independent
```

```python
# select the dependent varaible from the dataframe
dependent = dataset[['Salary']]
dependent
```

```python
# split train and test data
from sklearn.model_selection import train_test_split
```

```python
#  test_size used or split the data as based on the size. here 30% data
# function will return 4 values first 2 variable input of train and test data. next 2 variable  output of train and test data
x_train , x_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
x_train , y_train,  x_test ,y_test
```

```python
# Model creation
from sklearn.linear_model import LinearRegression
```

```python
# store the function to variable
regressor = LinearRegression()
```

```python
# regressor.fit(x_train,y_train)
# I'm getting the following warning after upgrading to version 1.0 of scikit-learn. So add .values in below lines
# calculated weight and bias
regressor.fit(x_train.values,y_train.values)
```

```python
# get the weight result
weight = regressor.coef_
weight
```

```python
# get he bias values
bias = regressor.intercept_
bias
```

```python
# evaluation metric
# passing the test data to test the model
y_pred = regressor.predict(x_test.values)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
# to check the model is good or bad it's arount 1 so the model is good if less than 1 it's bad model
r_score = r2_score(y_test,y_pred)
r_score
```

```python
# use this package to save the model
import pickle
```

```python
filename = "Finalized_model_linear.sav"
```

```python
# dump the result and save the result as .sav file. the data stored as binary
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Deployment phase
# Load the save model
loaded_model = pickle.load(open("Finalized_model_linear.sav",'rb'))
```
